# Logical-to-physical resource estimation lab

```python
import pandas as pd
from ft_ham_sim.resources import tfim_trotter_resources
from ft_ham_sim.surface_code import estimate_surface_code
```

```python
logical = tfim_trotter_resources(16, 100, order=1, synthesis_error_budget=2.5e-4)
pd.Series(logical.as_dict())
```

```python
logical_locations = logical.cnot_count + logical.t_count_proxy
physical = estimate_surface_code(
    logical_qubits=logical.logical_qubits,
    logical_locations=logical_locations,
    qec_error_budget=2.5e-4,
    physical_error_rate=1e-3,
)
pd.Series(physical.as_dict())
```

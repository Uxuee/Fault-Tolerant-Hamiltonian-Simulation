# Trotter convergence lab

```python
import matplotlib.pyplot as plt
from ft_ham_sim.hamiltonians import tfim_components
from ft_ham_sim.trotter import exact_unitary, trotter_unitary
from ft_ham_sim.metrics import operator_error
```

```python
n, t = 4, 1.0
h_zz, h_x = tfim_components(n, coupling=1.0, field=0.8)
exact = exact_unitary(h_zz+h_x, t)
steps = [1,2,4,8,16,32]
errors = {order: [operator_error(exact, trotter_unitary((h_zz,h_x), t, r, order)) for r in steps] for order in (1,2)}
errors
```

```python
for order, values in errors.items():
    plt.loglog(steps, values, marker="o", label=f"order {order}")
plt.xlabel("steps")
plt.ylabel("operator error")
plt.legend();
```
